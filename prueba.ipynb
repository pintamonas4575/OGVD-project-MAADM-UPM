```python
import findspark

findspark.init()
from pyspark import SparkContext
```

```python
sc = SparkContext("local[*]", "Name of the Program")
```

```python
text = sc.textFile('quijote.text')
```

```python
def m1(s):

    s=s.split()
    l =[(i,1)for i in s]
    return l
```

```python
c=text.flatMap(m1)
```

```python
c=c.reduceByKey(lambda x,y:x+y)
```

```python
c.take(5)
```

```python
a=c.collect()
```

```python
sorted(a,reverse=True,key=lambda y: y[1])
```
